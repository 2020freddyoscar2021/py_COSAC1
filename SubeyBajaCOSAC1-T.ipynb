```python
import pandas as pd
from datetime import date, time, datetime
```

```python
df1= pd.read_csv(r"Reporte_de_Validaciones_Diarias_de_Zona_Sur(4).csv", sep=",")
df2= pd.read_csv(r"Reporte_de_Validaciones_Diarias_de_Estacion_Central(4).csv", sep=",")
df3= pd.read_csv(r"Reporte_de_Validaciones_Diarias_de_Matellini(4).csv", sep=",")
df4= pd.read_csv(r"Reporte_de_Validaciones_Diarias_de_Naranjal(4).csv", sep=",")
df5= pd.read_csv(r"Reporte_de_Validaciones_Diarias_de_Zona_Centro(4).csv", sep=",")
df6= pd.read_csv(r"Reporte_de_Validaciones_Diarias_de_Zona_Norte(4).csv", sep=",")
df7= pd.read_csv(r"Reporte_de_Validaciones_Diarias_por_Estacion(4).csv", sep=",")
```

```python
df1_4Col= df1[['Numero Estacione','Estacione','Hora validacione','Numero de la tarjeta']]
df2_4Col= df2[['Numero Estacione','Estacione','Hora validacione','Numero de la tarjeta']]
df3_4Col= df3[['Numero Estacione','Estacione','Hora validacione','Numero de la tarjeta']]
df4_4Col= df4[['Numero Estacione','Estacione','Hora validacione','Numero de la tarjeta']]
df5_4Col= df5[['Numero Estacione','Estacione','Hora validacione','Numero de la tarjeta']]
df6_4Col= df6[['Numero Estacione','Estacione','Hora validacione','Numero de la tarjeta']]
df7_4Col= df7[['Numero Estacione','Estacione','Hora validacione','Numero de la tarjeta']]
```

```python
df_Union=[df1_4Col,df2_4Col,df3_4Col,df4_4Col,df5_4Col,df6_4Col,df7_4Col]
```

```python
df_General= pd.concat(df_Union)
```

```python
df_Gen_3Col=df_General[['Numero Estacione','Estacione','Numero de la tarjeta']]
```

```python
df_General['HORAN'] = df_General['Hora validacione'].str.extract('([0-9]+)')
df_General.head()
```

```python
df_General.groupby('HORAN').count()
```

```python
df_18pm = df_General[df_General['HORAN'] == '17']
df_18pm.head()
```

```python
Col_pivot=pd.pivot_table(df_18pm,index=['Numero Estacione','Estacione'], aggfunc='count')
Col_pivot.head()
```

```python
df_18menos = df_General[(df_General['HORAN'] < '17') | (df_General['HORAN'] > '17')]
df_18menos.head()
```

```python
df_18pm_df_18menos = pd.merge(df_18pm, df_18menos, on=['Numero de la tarjeta'], how='inner', indicator=True)
df_18pm_df_18menos
```

```python
df_18pm_df_18menos['LOGICO'] = ''
df_18pm_df_18menos.loc[df_18pm_df_18menos['Estacione_x'] == df_18pm_df_18menos['Estacione_y'], 'LOGICO'] = '0'
df_18pm_df_18menos.loc[df_18pm_df_18menos['Estacione_x'] != df_18pm_df_18menos['Estacione_y'], 'LOGICO'] = '1'
df_18pm_df_18menos.head()
```

```python
with pd.ExcelWriter('py_salidaCOSAC1_2019-06-14-(17pm)T.xlsx') as writer:
    df_18pm_df_18menos.to_excel(writer, index=False, sheet_name='basedatos')
    Col_pivot.to_excel(writer, index= True, sheet_name='estaciones')
```
